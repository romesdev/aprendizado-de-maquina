### 2.4. Crie um algoritmo seu (sem usar aprendizado de máquina) para baseado nas médias das features calculadas no item 2.3. ou mesmo usando outras ideias suas, classificar corretamente as amostras do conjunto de teste. Aperfeçoe o seu algoritmo de forma a obter melhores resultados. Quanto por cento no conjunto de teste você conseguiu acertar corretamente o rótulo (classe) no seu melhor algoritmo, ou seja, qual a acurácia?

A ideia implementada aqui é baseada nas médias das médias das features para cada Espécie. Então, dado a média dos dados de uma linha, o intervalo que essa média estiver próximo será a classe dessa linha. 

```python
def classificacao(data):
    resu = np.mean(data)
    
    if(resu < 3): return "Iris-setosa"
    if(resu > 3 and resu < 4): return "Iris-versicolor"
    if(resu > 4): return "Iris-virginica"
```

```python
r = []
for index, row in teste.iterrows(): 
    a = [row['SepalLengthCm'], row['SepalWidthCm'], row['PetalLengthCm'], row['PetalWidthCm']]
    r.append([index, classificacao(a)])
```

```python
lista = teste['Species']

def acuracia():
    certos = 0
    count = 0
    for i, k in zip(r, lista):
        count += 1
        print(count)
        print("prev: ", i[1], " real: ", k)
        if(i[1] == k):
            certos += 1
        
        print("certos: ", certos)
    
    acu = certos/len(lista)
    return acu
```

```python
acu = acuracia()
```

```python
print("Acurácia da Classificação: ", round(acu * 100, 2))
```
